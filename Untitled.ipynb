```python
import numpy as np
from scipy import stats

SG =np.array([16.3,11.0,-2.6,6.0, 15.4,5.1,5.1,7.0,9.4,-6.7])
SG_F =np.array([12.2, -1.5, 7.3, 16.5, 6.1, 6.2, 8.1, 10.5,-5.7])
SGRS=np.array([7.8,-4.5,8.4,16.4,6.9,6.2,4.2,7.4,-4.2])


```

```python
stats.describe(SG)
```

```python
stats.describe(SGRS)
```

```python
stats.describe(SG_F)
```

```python
np.quantile(SG,[0,0.25,0.5,0.75,1.0])
```

```python
np.quantile(SGRS,[0,0.25,0.5,0.75,1.0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# matplotlib histogram
## seaborn histogram
sns.distplot(SG, hist=True, kde=True, color = 'blue', bins=20, hist_kws={'edgecolor':'black'},label='SG')
#sns.distplot(SGRS, hist=True, kde=True, color = 'red',bins=20, hist_kws={'edgecolor':'black'},label='SGRS')
sns.distplot(SG_F, hist=True, kde=True, color = 'green',bins=20, hist_kws={'edgecolor':'black'},label='SG_F')

# Add labels
plt.title('distribution of profit')
plt.xlabel('profit')
plt.ylabel('probability')
plt.legend()
plt.show()

```

```python


kwargs = {'cumulative': True}
sns.distplot(SG, hist_kws=kwargs, bins=20,color='blue', label='SG')
#sns.distplot(SGRS, hist_kws=kwargs, bins=20,color='red',  label='SGRS')
sns.distplot(SG_F, hist_kws=kwargs, bins=20,color='green',  label='SG_F')

# # Add labels
plt.title('distribution of profit')
plt.xlabel('profit')
plt.ylabel('cdf')
plt.legend()
plt.grid()
plt.show()

```
